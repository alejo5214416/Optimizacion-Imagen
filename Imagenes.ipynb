### Paso 2: Propiedades de la imagen

```python
%pylab inline  
import skimage.io as skio

image = skio.imread('3.jpg')
plt.imshow(image)
```

La imagen contiene un vector de tres dimensiones. Una para la coordenada de x, una segunda coordenada para y, ambas coordenadas de la ubicacion de cada pixel, y una tercera que dá informacion sobre la composicion del color de dicho pixel en RGB, tambien un cuarto factor alpha para la transparencia. Por ejemplo

     I(x,y,r,g,b) = (100, 200, 0.2, 0.1, 0.8, 0.9)


```python
'shape = %s   dimensions = %s' % (image.shape, image.ndim)
```

### Paso 3: 1 Canal-Escala de grises

Es posible convertir la imagen de 3 canales a una imagen de 1 canal sin brillo ( Espacio de escala de grises). Scikit-Image realiza estas conversiones con su modulo skcolor.

```python
import skimage.color as skcolor

grayimage = skcolor.rgb2gray(image)
plt.imshow(grayimage)
plt.title('Imagen 1 canal');
'shape = %s   dimensions = %s' % (grayimage.shape, grayimage.ndim)
```

```python
plt.imshow(grayimage, cmap='gray')
plt.title('Imagen de 1 Canal mapeada enn gris');
```

### Histograma de brillo de la imagen

```python
plt.hist(grayimage)
plt.xlabel('Brightness (0=negro, 1 = blanco)')
plt.ylabel('Conteo')
plt.title('Distribucion de brillo de pixeles');
```
